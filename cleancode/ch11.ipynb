# 주석 설명

  - `TODO` : 해야 할 일에 대한 내용을 일반저그로 상기시켜야 하는 경우

# 타입 힌트를 다중 타입으로 설정 

````python
from typing import Union

sapm : Union[int, str, float] = 42
print(sapm)
sapm = 'spam'
print(sapm)
sapm = 3.14
print(sapm)
````

````python
# Union [ str, None ]을 사용하기 보단 
# Optional [ str ]을 사용하는 것이 더 간결하다.
from typing import Optional
bacon: Optional[str] = None
````

````python
# Any Type
from typing import Any 
import datetime
spam: Any = 42
spam = datetime.date.today()
spam = True 
````

# List, Dictionary 등을 위한 타입 힌트 설정 


````python
# 리스트는 모든 타입의 값을 포함할 수 있다. 
from typing import List, Union
catNames: List[str] = [ 'Zopie', 'Pooka', 'Simon' ]
numbers : List[Union[int, float]] = [ 1, 2, 3.14, 4, 5.0 ]
````

# typing module은 별도의 타입 별칭(type alias)를 가진다. 

- `List` : List datetype
- `Tuple` : tuple data type 
- `Dict` : Dictionary data type 
- `Set` : Set data type
- `FrozenSet` : frozenset data type
- `Sequence` : list, tuple, etc sequence type
- `Mapping` : dict, set, frozenset, etc mapping type
- `ByteString` : byte, bytearray, memoryview data type
