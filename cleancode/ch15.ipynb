# isinstance(), issubclass()



```python
class ParentClass:
    pass
class ChildClass(ParentClass):
    pass

isinstance(ChildClass(), ParentClass)
```

```python
isinstance(42, int)
```

```python
isinstance(ParentClass(), ChildClass)
```

```python
isinstance(ChildClass(), ChildClass)
```

```python
isinstance(ChildClass(), ParentClass)
```

```python
isinstance(42, (int, str, float ))
```

```python
issubclass(ChildClass, ParentClass)
```

```python
issubclass(ChildClass, str)
```

```python
issubclass(ChildClass,ChildClass)
```

# class method

```python
class ExampleClass:
    def exampleRegularMethod(self):
        print("this is a regular method")
    @classmethod
    def exampleClassMethod(cls):
        print("this is class method of", cls.__name__)
    @staticmethod
    def exampleStaticMethod():
        print("this is a static method")

ExampleClass.exampleStaticMethod()
```

```python
ExampleClass.exampleClassMethod()
```

```python
obj = ExampleClass()
obj.exampleRegularMethod()
obj.__class__.exampleClassMethod()
```

- self 는  객체를 참조 
- cls파라미터는 객체의 클래스를 참조 
- class method : [classmethod](./classmethod.md)

```python
class AsciiArt:
    def __init__(self, characters):
        self._characters = characters
    @classmethod
    def fromFile(cls, filename):
        with open(filename) as f:
            characters = f.read()
            return cls(characters)
    def display(self):
        print(self._characters)
```

```python
# class의 속성 

class CreateCount:
    count = 0 #class의 속성 
    def __init__(self):
        self.count_value = 0
        CreateCount.count += 1

print('생성된 객체:', CreateCount.count)
a = CreateCount()
b = CreateCount()
c = CreateCount()
print('생성된 객체:', CreateCount.count, a.count_value)
```

```python
#static method 
class ExampleCalssWithStaticMethod:
    @staticmethod
    def add(a, b):
        return a + b
c = ExampleCalssWithStaticMethod.add(3, 5)
print(c)
```

```python
class test:
    def __init__(self):
        self._private_value = 42
test_instance = test()
print(test_instance._private_value)
```
