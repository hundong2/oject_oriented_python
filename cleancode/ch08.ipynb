# loop진행 중에는 리스트에서 아이템을 추가, 삭제하지말자

````python
clothes = ['skirt', 'red socks' ]
for cloth in clothes:
    if 'sock' in cloth:
        clothes.append(cloth)
        print('Added a sock:', cloth)
````

````python
clothes = ['skirt', 'red socks' ]
new_clothes = clothes.copy()
for cloth in clothes:
    if 'sock' in cloth:
        new_clothes.append(cloth)
        print('Added a sock:', cloth)
````
