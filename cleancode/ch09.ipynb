# 파이썬의 기이한 특징

## python

- 파이썬은 프로그램 시작 시 `-5` ~ `256`까지의 정수객체를 미리 생성해둔다. 

```python
a = 42
b = 42.0
a == b 
```

```python
a is b 
```

```python
a = 256
b = 256
a is b
```

```python
a = 257
b = 257
a is b
```

```python
257 is 257 
#C 파이썬에서 하나의 문(statement)안에서
# 특정 리터럴에 대응하는 정수 객체를 다른 리터럴에서도 재사용 
```

# 최적화 ( 문자열 인터닝 )

```python
spam = 'cat'
eggs = 'cat'
print(spam is eggs)  # True
print(id(spam), id(eggs))  # 서로 같은 id 값 출력
```

# 아무것도 없으면 참인 all()

- `all() 은 하나이상의 값이 False이면 False`

```python
spam = [67, 39, 20, 55, 13, 45, 44]
all([i > 42 for i in spam])
```

```python
all([]) #이 리스트의 어느것도 False 가 아니다. 이므로 True
```
