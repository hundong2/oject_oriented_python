## 1. í´ë˜ìŠ¤ ì¸ìŠ¤í„´ìŠ¤ ìƒì„± ê³¼ì •

Pythonì—ì„œ `MyClass()`ë¥¼ í˜¸ì¶œí•  ë•Œ ì‹¤ì œë¡œ ì¼ì–´ë‚˜ëŠ” ê³¼ì •:

```
1. __new__() í˜¸ì¶œ â†’ ë©”ëª¨ë¦¬ì— ê°ì²´ ìƒì„±
2. __init__() í˜¸ì¶œ â†’ ê°ì²´ ì´ˆê¸°í™”
3. ì´ˆê¸°í™”ëœ ê°ì²´ ë°˜í™˜
```

### ê¸°ë³¸ ì˜ˆì œë¡œ ì´í•´í•˜ê¸°

````python
class BasicExample:
    def __new__(cls, *args, **kwargs):
        print(f"1. __new__ í˜¸ì¶œë¨: {cls}")
        print(f"   ì¸ìˆ˜: args={args}, kwargs={kwargs}")
        # object.__new__()ë¥¼ í˜¸ì¶œí•˜ì—¬ ì‹¤ì œ ì¸ìŠ¤í„´ìŠ¤ ìƒì„±
        instance = super().__new__(cls)
        print(f"   ìƒì„±ëœ ì¸ìŠ¤í„´ìŠ¤: {instance}")
        return instance
    
    def __init__(self, name, value):
        print(f"2. __init__ í˜¸ì¶œë¨: {self}")
        print(f"   ì¸ìˆ˜: name={name}, value={value}")
        self.name = name
        self.value = value
        print(f"   ì´ˆê¸°í™” ì™„ë£Œ")

print("=== ê°ì²´ ìƒì„± ê³¼ì • ê´€ì°° ===")
obj = BasicExample("í…ŒìŠ¤íŠ¸", 42)
print(f"\nìµœì¢… ê°ì²´: {obj}")
print(f"ê°ì²´ ì†ì„±: name={obj.name}, value={obj.value}")
````

## 2. `__new__()` ë©”ì„œë“œ ì‹¬í™” ì´í•´

### `__new__()`ì˜ íŠ¹ì§•
- **ì •ì  ë©”ì„œë“œ**: ì²« ë²ˆì§¸ ì¸ìˆ˜ë¡œ í´ë˜ìŠ¤(`cls`)ë¥¼ ë°›ìŒ
- **ê°ì²´ ìƒì„±**: ì‹¤ì œ ë©”ëª¨ë¦¬ì— ê°ì²´ë¥¼ ìƒì„±í•˜ëŠ” ì—­í• 
- **ë°˜í™˜ê°’**: ìƒì„±ëœ ì¸ìŠ¤í„´ìŠ¤ë¥¼ ë°˜í™˜í•´ì•¼ í•¨
- **í˜¸ì¶œ ì‹œì **: `__init__()` ë³´ë‹¤ ë¨¼ì € í˜¸ì¶œë¨

### `__new__()`ê°€ í•„ìš”í•œ ê²½ìš°
1. **ë¶ˆë³€ ê°ì²´** (int, str, tuple ë“±) ìƒì† ì‹œ
2. **ì‹±ê¸€í†¤ íŒ¨í„´** êµ¬í˜„ ì‹œ
3. **ê°ì²´ ìƒì„±ì„ ì œì–´**í•´ì•¼ í•  ë•Œ

````python
# ì˜ˆì œ 1: ì‹±ê¸€í†¤ íŒ¨í„´ êµ¬í˜„
class Singleton:
    _instance = None
    
    def __new__(cls):
        if cls._instance is None:
            print("ìƒˆë¡œìš´ ì‹±ê¸€í†¤ ì¸ìŠ¤í„´ìŠ¤ ìƒì„±")
            cls._instance = super().__new__(cls)
        else:
            print("ê¸°ì¡´ ì‹±ê¸€í†¤ ì¸ìŠ¤í„´ìŠ¤ ë°˜í™˜")
        return cls._instance
    
    def __init__(self):
        # ì¤‘ë³µ ì´ˆê¸°í™” ë°©ì§€
        if not hasattr(self, 'initialized'):
            print("ì‹±ê¸€í†¤ ì´ˆê¸°í™”")
            self.data = "ì‹±ê¸€í†¤ ë°ì´í„°"
            self.initialized = True

print("=== ì‹±ê¸€í†¤ íŒ¨í„´ í…ŒìŠ¤íŠ¸ ===")
s1 = Singleton()
s2 = Singleton()
s3 = Singleton()

print(f"s1 == s2: {s1 == s2}")
print(f"s1 is s2: {s1 is s2}")
print(f"id(s1): {id(s1)}")
print(f"id(s2): {id(s2)}")
print(f"id(s3): {id(s3)}")
````

## 3. ë¶ˆë³€ ê°ì²´ì—ì„œì˜ `__new__()` ì‚¬ìš©

ë¶ˆë³€ ê°ì²´(int, str, tuple ë“±)ë¥¼ ìƒì†í•  ë•ŒëŠ” `__init__()`ë§Œìœ¼ë¡œëŠ” ë¶€ì¡±í•©ë‹ˆë‹¤.
ì™œëƒí•˜ë©´ ë¶ˆë³€ ê°ì²´ëŠ” ìƒì„± í›„ ìˆ˜ì •í•  ìˆ˜ ì—†ê¸° ë•Œë¬¸ì…ë‹ˆë‹¤.

### ë¬¸ì œê°€ ìˆëŠ” ì½”ë“œ (ch03_ex5.pyì—ì„œ ë°œì·Œ)

````python
# ë¬¸ì œê°€ ìˆëŠ” ì½”ë“œ: floatë¥¼ ìƒì†í•˜ë©´ì„œ __init__ë§Œ ì‚¬ìš©
class Float_Fail(float):
    def __init__(self, value: float, unit: str) -> None:
        super().__init__(value)  # ì´ ë¶€ë¶„ì´ ë¬¸ì œ!
        self.unit = unit

print("=== ì˜ëª»ëœ ë°©ë²• í…ŒìŠ¤íŠ¸ ===")
try:
    x = Float_Fail(6.8, "knots")
    print(f"ì„±ê³µ: {x}, ë‹¨ìœ„: {x.unit}")
except Exception as e:
    print(f"ì˜¤ë¥˜ ë°œìƒ: {type(e).__name__}: {e}")
    print("ì™œ ì˜¤ë¥˜ê°€ ë‚ ê¹Œìš”? floatëŠ” ë¶ˆë³€ ê°ì²´ì´ë¯€ë¡œ __init__ì—ì„œ ê°’ì„ ì„¤ì •í•  ìˆ˜ ì—†ìŠµë‹ˆë‹¤!")
````

````python
# ì˜¬ë°”ë¥¸ ë°©ë²•: __new__ë¥¼ ì‚¬ìš©í•˜ì—¬ ë¶ˆë³€ ê°ì²´ í™•ì¥
class Float_Units(float):
    def __new__(cls, value, unit):
        print(f"Float_Units.__new__ í˜¸ì¶œ: value={value}, unit={unit}")
        
        # 1. ë¶€ëª¨ í´ë˜ìŠ¤ì˜ __new__ë¡œ float ê°ì²´ ìƒì„±
        obj = super().__new__(cls, float(value))
        print(f"ìƒì„±ëœ float ê°ì²´: {obj} (íƒ€ì…: {type(obj)})")
        
        # 2. ì¶”ê°€ ì†ì„±ì„ ê°ì²´ì— ì„¤ì • (ë¶ˆë³€ ê°ì²´ê°€ ìƒì„±ëœ ì§í›„ì´ë¯€ë¡œ ê°€ëŠ¥)
        obj.unit = unit
        print(f"ë‹¨ìœ„ ì„¤ì • ì™„ë£Œ: {obj.unit}")
        
        return obj

print("=== ì˜¬ë°”ë¥¸ ë°©ë²• í…ŒìŠ¤íŠ¸ ===")
speed = Float_Units(6.8, "knots")
print(f"ìƒì„±ëœ ê°ì²´: {speed}")
print(f"íƒ€ì…: {type(speed)}")
print(f"ê°’: {speed}")
print(f"ë‹¨ìœ„: {speed.unit}")
print(f"ì—°ì‚° ê°€ëŠ¥: {speed * 2}")
print(f"floatì²˜ëŸ¼ ë™ì‘: {speed + 3.2}")
````

````python
# ë˜ ë‹¤ë¥¸ ì˜ˆì œ: ë¬¸ìì—´ í™•ì¥
class UpperString(str):
    def __new__(cls, value):
        # ë¬¸ìì—´ì„ ëŒ€ë¬¸ìë¡œ ë³€í™˜í•˜ì—¬ ìƒì„±
        obj = super().__new__(cls, value.upper())
        obj.original = value  # ì›ë³¸ ê°’ ë³´ê´€
        return obj
    
    def get_original(self):
        return self.original

print("=== ë¬¸ìì—´ í™•ì¥ ì˜ˆì œ ===")
text = UpperString("Hello World")
print(f"ìƒì„±ëœ ë¬¸ìì—´: '{text}'")
print(f"ì›ë³¸: '{text.get_original()}'")
print(f"íƒ€ì…: {type(text)}")
print(f"str ë©”ì„œë“œ ì‚¬ìš© ê°€ëŠ¥: '{text.lower()}'")

# ë¹„êµ: ì¼ë°˜ strê³¼ì˜ ì°¨ì´
normal_str = str("Hello World")
print(f"\nì¼ë°˜ str: '{normal_str}'")
print(f"ìš°ë¦¬ í´ë˜ìŠ¤: '{text}'")
````

## 4. ë©”íƒ€í´ë˜ìŠ¤ì™€ `__new__()`

ë©”íƒ€í´ë˜ìŠ¤ëŠ” "í´ë˜ìŠ¤ë¥¼ ë§Œë“œëŠ” í´ë˜ìŠ¤"ì…ë‹ˆë‹¤. 
ë©”íƒ€í´ë˜ìŠ¤ì˜ `__new__()`ëŠ” í´ë˜ìŠ¤ ìì²´ë¥¼ ìƒì„±í•  ë•Œ í˜¸ì¶œë©ë‹ˆë‹¤.

### ë©”íƒ€í´ë˜ìŠ¤ ë™ì‘ ê³¼ì •
1. í´ë˜ìŠ¤ ì •ì˜ë¥¼ ë§Œë‚¨
2. ë©”íƒ€í´ë˜ìŠ¤ì˜ `__new__()` í˜¸ì¶œ â†’ í´ë˜ìŠ¤ ê°ì²´ ìƒì„±
3. í´ë˜ìŠ¤ì˜ `__new__()` í˜¸ì¶œ â†’ ì¸ìŠ¤í„´ìŠ¤ ìƒì„±
4. í´ë˜ìŠ¤ì˜ `__init__()` í˜¸ì¶œ â†’ ì¸ìŠ¤í„´ìŠ¤ ì´ˆê¸°í™”

````python
import logging
from typing import Dict, Any, Tuple, Type

# ë©”íƒ€í´ë˜ìŠ¤ ì˜ˆì œ: ìë™ ë¡œê±° ì¶”ê°€
class LoggedMeta(type):
    def __new__(cls, name: str, bases: Tuple[Type, ...], namespace: Dict[str, Any]):
        print(f"ë©”íƒ€í´ë˜ìŠ¤ __new__ í˜¸ì¶œ: í´ë˜ìŠ¤ '{name}' ìƒì„± ì¤‘...")
        
        # í´ë˜ìŠ¤ ìƒì„±
        result = super().__new__(cls, name, bases, namespace)
        
        # ìƒì„±ëœ í´ë˜ìŠ¤ì— ìë™ìœ¼ë¡œ ë¡œê±° ì¶”ê°€
        result.logger = logging.getLogger(name)
        print(f"ë¡œê±° ì¶”ê°€ ì™„ë£Œ: {result.logger}")
        
        return result

class Logged(metaclass=LoggedMeta):
    """ë¡œê±°ê°€ ìë™ìœ¼ë¡œ ì¶”ê°€ë˜ëŠ” ë² ì´ìŠ¤ í´ë˜ìŠ¤"""
    pass

print("=== ë©”íƒ€í´ë˜ìŠ¤ í…ŒìŠ¤íŠ¸ ===")

class MyApplication(Logged):
    def __init__(self, name):
        self.name = name
        self.logger.info(f"MyApplication ìƒì„±: {name}")
    
    def do_something(self):
        self.logger.info("ë­”ê°€ ì¤‘ìš”í•œ ì‘ì—… ìˆ˜í–‰ ì¤‘...")
        return "ì‘ì—… ì™„ë£Œ"

# ë¡œê¹… ì„¤ì •
logging.basicConfig(level=logging.INFO, format='%(name)s - %(levelname)s - %(message)s')

print("\n=== ì• í”Œë¦¬ì¼€ì´ì…˜ ì‹¤í–‰ ===")
app = MyApplication("í…ŒìŠ¤íŠ¸ ì•±")
result = app.do_something()
print(f"ê²°ê³¼: {result}")

# ë¡œê±°ê°€ ìë™ìœ¼ë¡œ ì¶”ê°€ë˜ì—ˆëŠ”ì§€ í™•ì¸
print(f"\në¡œê±° í™•ì¸: {hasattr(MyApplication, 'logger')}")
print(f"ë¡œê±° ì´ë¦„: {MyApplication.logger.name}")
````

## 5. ê³ ê¸‰ ê°ì²´ ìƒì„± ì œì–´ ì˜ˆì œ

### ì˜ˆì œ 1: íŒ©í† ë¦¬ íŒ¨í„´ êµ¬í˜„

````python
# ì˜ˆì œ 1: ì¸ìŠ¤í„´ìŠ¤ ê°œìˆ˜ ì œí•œ
class LimitedInstances:
    _instances = []
    _max_instances = 3
    
    def __new__(cls, name):
        if len(cls._instances) >= cls._max_instances:
            print(f"ìµœëŒ€ {cls._max_instances}ê°œì˜ ì¸ìŠ¤í„´ìŠ¤ë§Œ í—ˆìš©ë©ë‹ˆë‹¤!")
            print("ê¸°ì¡´ ì¸ìŠ¤í„´ìŠ¤ ì¤‘ í•˜ë‚˜ë¥¼ ë°˜í™˜í•©ë‹ˆë‹¤.")
            return cls._instances[0]  # ì²« ë²ˆì§¸ ì¸ìŠ¤í„´ìŠ¤ ì¬ì‚¬ìš©
        
        print(f"ìƒˆ ì¸ìŠ¤í„´ìŠ¤ ìƒì„± ì¤‘... (í˜„ì¬: {len(cls._instances)}ê°œ)")
        instance = super().__new__(cls)
        cls._instances.append(instance)
        return instance
    
    def __init__(self, name):
        if hasattr(self, 'name'):
            print(f"ì´ë¯¸ ì´ˆê¸°í™”ëœ ì¸ìŠ¤í„´ìŠ¤: {self.name}")
            return
        
        self.name = name
        print(f"ì¸ìŠ¤í„´ìŠ¤ ì´ˆê¸°í™”: {name}")

print("=== ì¸ìŠ¤í„´ìŠ¤ ê°œìˆ˜ ì œí•œ í…ŒìŠ¤íŠ¸ ===")
obj1 = LimitedInstances("ì²«ë²ˆì§¸")
obj2 = LimitedInstances("ë‘ë²ˆì§¸") 
obj3 = LimitedInstances("ì„¸ë²ˆì§¸")
obj4 = LimitedInstances("ë„¤ë²ˆì§¸")  # ì œí•œ ì´ˆê³¼
obj5 = LimitedInstances("ë‹¤ì„¯ë²ˆì§¸")  # ì œí•œ ì´ˆê³¼

print(f"\nobj1 == obj4: {obj1 is obj4}")
print(f"obj1.name: {obj1.name}")
print(f"obj4.name: {obj4.name}")
print(f"ì´ ì¸ìŠ¤í„´ìŠ¤ ê°œìˆ˜: {len(LimitedInstances._instances)}")
````

````python
# ì˜ˆì œ 2: ì¡°ê±´ë¶€ ê°ì²´ ìƒì„±
class ConditionalObject:
    def __new__(cls, value):
        if value < 0:
            print(f"ìŒìˆ˜ ê°’ {value}ì€ í—ˆìš©ë˜ì§€ ì•ŠìŠµë‹ˆë‹¤. Noneì„ ë°˜í™˜í•©ë‹ˆë‹¤.")
            return None
        
        if value > 100:
            print(f"í° ê°’ {value}ì„ ì‘ì€ ê°’ìœ¼ë¡œ ì œí•œí•©ë‹ˆë‹¤.")
            instance = super().__new__(cls)
            instance._value = 100
            return instance
        
        print(f"ì •ìƒ ê°’ {value}ìœ¼ë¡œ ê°ì²´ë¥¼ ìƒì„±í•©ë‹ˆë‹¤.")
        instance = super().__new__(cls)
        instance._value = value
        return instance
    
    def __init__(self, value):
        # __new__ì—ì„œ Noneì„ ë°˜í™˜í•˜ë©´ __init__ì€ í˜¸ì¶œë˜ì§€ ì•ŠìŒ
        print(f"__init__ í˜¸ì¶œ: value={value}, _value={getattr(self, '_value', 'N/A')}")
    
    def get_value(self):
        return getattr(self, '_value', None)

print("=== ì¡°ê±´ë¶€ ê°ì²´ ìƒì„± í…ŒìŠ¤íŠ¸ ===")

# ì •ìƒì ì¸ ê²½ìš°
obj1 = ConditionalObject(50)
print(f"obj1: {obj1}, ê°’: {obj1.get_value() if obj1 else 'None'}")

# ìŒìˆ˜ì¸ ê²½ìš° (None ë°˜í™˜)
obj2 = ConditionalObject(-10)
print(f"obj2: {obj2}")

# í° ê°’ì¸ ê²½ìš° (ì œí•œë¨)
obj3 = ConditionalObject(150)
print(f"obj3: {obj3}, ê°’: {obj3.get_value() if obj3 else 'None'}")
````

## 6. `cls` vs `self`: í•µì‹¬ ê°œë… ì´í•´

### ê¸°ë³¸ ê°œë…
- **`cls`**: í´ë˜ìŠ¤ ìì²´ë¥¼ ê°€ë¦¬í‚¤ëŠ” ë§¤ê°œë³€ìˆ˜ (Classì˜ ì¤„ì„ë§)
- **`self`**: ì¸ìŠ¤í„´ìŠ¤ ìì²´ë¥¼ ê°€ë¦¬í‚¤ëŠ” ë§¤ê°œë³€ìˆ˜ (Instanceì˜ ëŒ€ëª…ì‚¬)

### ì–¸ì œ ì–´ë–¤ ê²ƒì„ ì‚¬ìš©í•˜ë‚˜?

| ë©”ì„œë“œ ì¢…ë¥˜ | ë§¤ê°œë³€ìˆ˜ | ì—­í•  | í˜¸ì¶œ ì‹œì  |
|-------------|----------|------|-----------|
| `__new__()` | `cls` | í´ë˜ìŠ¤ë¡œë¶€í„° ì¸ìŠ¤í„´ìŠ¤ ìƒì„± | ì¸ìŠ¤í„´ìŠ¤ ìƒì„± ì „ |
| `__init__()` | `self` | ìƒì„±ëœ ì¸ìŠ¤í„´ìŠ¤ ì´ˆê¸°í™” | ì¸ìŠ¤í„´ìŠ¤ ìƒì„± í›„ |
| `@classmethod` | `cls` | í´ë˜ìŠ¤ ì°¨ì›ì˜ ì‘ì—… | ì–¸ì œë“ ì§€ |
| ì¼ë°˜ ë©”ì„œë“œ | `self` | ì¸ìŠ¤í„´ìŠ¤ ì°¨ì›ì˜ ì‘ì—… | ì¸ìŠ¤í„´ìŠ¤ ìƒì„± í›„ |

````python
# clsì™€ self ì°¨ì´ì  ì‹¤ìŠµ
class ClsVsSelfDemo:
    class_variable = "í´ë˜ìŠ¤ ë³€ìˆ˜"
    instance_count = 0
    
    def __new__(cls, name):
        print(f"=== __new__ ë©”ì„œë“œ ===")
        print(f"clsëŠ” í´ë˜ìŠ¤ ìì²´: {cls}")
        print(f"cls.__name__: {cls.__name__}")
        print(f"cls.class_variable: {cls.class_variable}")
        
        # clsë¡œ í´ë˜ìŠ¤ ë³€ìˆ˜ì— ì ‘ê·¼
        cls.instance_count += 1
        print(f"ìƒì„±ëœ ì¸ìŠ¤í„´ìŠ¤ ê°œìˆ˜: {cls.instance_count}")
        
        # ì¸ìŠ¤í„´ìŠ¤ ìƒì„±
        instance = super().__new__(cls)
        print(f"ìƒì„±ëœ ì¸ìŠ¤í„´ìŠ¤: {instance}")
        return instance
    
    def __init__(self, name):
        print(f"\n=== __init__ ë©”ì„œë“œ ===")
        print(f"selfëŠ” ì¸ìŠ¤í„´ìŠ¤ ìì²´: {self}")
        print(f"self.__class__: {self.__class__}")
        print(f"self.__class__.__name__: {self.__class__.__name__}")
        
        # selfë¡œ ì¸ìŠ¤í„´ìŠ¤ ë³€ìˆ˜ ì„¤ì •
        self.name = name
        self.instance_variable = f"{name}ì˜ ì¸ìŠ¤í„´ìŠ¤ ë³€ìˆ˜"
        print(f"ì¸ìŠ¤í„´ìŠ¤ ë³€ìˆ˜ ì„¤ì •: {self.instance_variable}")
    
    @classmethod
    def class_method_example(cls):
        print(f"\n=== @classmethod ì˜ˆì œ ===")
        print(f"clsëŠ” í´ë˜ìŠ¤: {cls}")
        print(f"í´ë˜ìŠ¤ ë³€ìˆ˜ ì ‘ê·¼: {cls.class_variable}")
        print(f"í˜„ì¬ ì¸ìŠ¤í„´ìŠ¤ ê°œìˆ˜: {cls.instance_count}")
        return f"{cls.__name__}ì˜ í´ë˜ìŠ¤ ë©”ì„œë“œ ì‹¤í–‰"
    
    def instance_method_example(self):
        print(f"\n=== ì¸ìŠ¤í„´ìŠ¤ ë©”ì„œë“œ ì˜ˆì œ ===")
        print(f"selfëŠ” ì¸ìŠ¤í„´ìŠ¤: {self}")
        print(f"ì¸ìŠ¤í„´ìŠ¤ ë³€ìˆ˜ ì ‘ê·¼: {self.name}")
        print(f"í´ë˜ìŠ¤ ë³€ìˆ˜ì—ë„ ì ‘ê·¼ ê°€ëŠ¥: {self.class_variable}")
        return f"{self.name}ì˜ ì¸ìŠ¤í„´ìŠ¤ ë©”ì„œë“œ ì‹¤í–‰"

print("=== ì²« ë²ˆì§¸ ê°ì²´ ìƒì„± ===")
obj1 = ClsVsSelfDemo("ê°ì²´1")

print("\n=== ë‘ ë²ˆì§¸ ê°ì²´ ìƒì„± ===") 
obj2 = ClsVsSelfDemo("ê°ì²´2")

print("\n=== í´ë˜ìŠ¤ ë©”ì„œë“œ í˜¸ì¶œ ===")
result = ClsVsSelfDemo.class_method_example()
print(f"ê²°ê³¼: {result}")

print("\n=== ì¸ìŠ¤í„´ìŠ¤ ë©”ì„œë“œ í˜¸ì¶œ ===")
result1 = obj1.instance_method_example()
result2 = obj2.instance_method_example()
print(f"obj1 ê²°ê³¼: {result1}")
print(f"obj2 ê²°ê³¼: {result2}")
````

### clsì™€ selfì˜ ì ‘ê·¼ ë²”ìœ„

#### `cls`ë¡œ ì ‘ê·¼ ê°€ëŠ¥í•œ ê²ƒë“¤:
- âœ… í´ë˜ìŠ¤ ë³€ìˆ˜
- âœ… í´ë˜ìŠ¤ ë©”ì„œë“œ (`@classmethod`)
- âœ… ì •ì  ë©”ì„œë“œ (`@staticmethod`)
- âœ… ë‹¤ë¥¸ í´ë˜ìŠ¤ë“¤ (ìƒì† ê´€ê³„)
- âŒ ì¸ìŠ¤í„´ìŠ¤ ë³€ìˆ˜ (ì•„ì§ ìƒì„±ë˜ì§€ ì•ŠìŒ)

#### `self`ë¡œ ì ‘ê·¼ ê°€ëŠ¥í•œ ê²ƒë“¤:
- âœ… ì¸ìŠ¤í„´ìŠ¤ ë³€ìˆ˜
- âœ… ì¸ìŠ¤í„´ìŠ¤ ë©”ì„œë“œ
- âœ… í´ë˜ìŠ¤ ë³€ìˆ˜ (ìƒì†ì„ í†µí•´)
- âœ… í´ë˜ìŠ¤ ë©”ì„œë“œ (ìƒì†ì„ í†µí•´)

### ì‹¤ì œ í™œìš© ì‚¬ë¡€

````python
# ì‹¤ì œ í™œìš© ì‚¬ë¡€: ë‹¤ì–‘í•œ ìƒì„±ì íŒ¨í„´
class Person:
    population = 0  # í´ë˜ìŠ¤ ë³€ìˆ˜
    
    def __new__(cls, *args, **kwargs):
        print(f"Person.__new__ í˜¸ì¶œ, cls={cls.__name__}")
        cls.population += 1  # clsë¡œ í´ë˜ìŠ¤ ë³€ìˆ˜ ìˆ˜ì •
        return super().__new__(cls)
    
    def __init__(self, name, age):
        print(f"Person.__init__ í˜¸ì¶œ, self={id(self)}")
        self.name = name    # selfë¡œ ì¸ìŠ¤í„´ìŠ¤ ë³€ìˆ˜ ì„¤ì •
        self.age = age      # selfë¡œ ì¸ìŠ¤í„´ìŠ¤ ë³€ìˆ˜ ì„¤ì •
    
    @classmethod
    def from_birth_year(cls, name, birth_year):
        """clsë¥¼ ì‚¬ìš©í•œ ëŒ€ì•ˆ ìƒì„±ì"""
        print(f"from_birth_year í˜¸ì¶œ, cls={cls.__name__}")
        current_year = 2024
        age = current_year - birth_year
        # cls()ë¥¼ í˜¸ì¶œí•˜ì—¬ ì¸ìŠ¤í„´ìŠ¤ ìƒì„±
        return cls(name, age)
    
    @classmethod
    def get_population(cls):
        """clsë¥¼ ì‚¬ìš©í•œ í´ë˜ìŠ¤ ì •ë³´ ì¡°íšŒ"""
        return cls.population
    
    def get_info(self):
        """selfë¥¼ ì‚¬ìš©í•œ ì¸ìŠ¤í„´ìŠ¤ ì •ë³´ ì¡°íšŒ"""
        return f"{self.name} ({self.age}ì„¸)"
    
    def have_birthday(self):
        """selfë¥¼ ì‚¬ìš©í•œ ì¸ìŠ¤í„´ìŠ¤ ìƒíƒœ ë³€ê²½"""
        self.age += 1
        return f"{self.name}ë‹˜, ìƒì¼ ì¶•í•˜í•©ë‹ˆë‹¤! ì´ì œ {self.age}ì„¸ì…ë‹ˆë‹¤."

print("=== clsë¥¼ í™œìš©í•œ ë‹¤ì–‘í•œ ìƒì„± ë°©ë²• ===")

# ì¼ë°˜ì ì¸ ìƒì„±ì ì‚¬ìš©
person1 = Person("ê¹€ì² ìˆ˜", 25)
print(f"ìƒì„±: {person1.get_info()}")

# clsë¥¼ í™œìš©í•œ ëŒ€ì•ˆ ìƒì„±ì ì‚¬ìš©
person2 = Person.from_birth_year("ì´ì˜í¬", 1995)
print(f"ìƒì„±: {person2.get_info()}")

# clsë¥¼ í™œìš©í•œ í´ë˜ìŠ¤ ì •ë³´ ì¡°íšŒ
print(f"ì´ ì¸êµ¬: {Person.get_population()}ëª…")

print("\n=== selfë¥¼ í™œìš©í•œ ì¸ìŠ¤í„´ìŠ¤ ì¡°ì‘ ===")
print(person1.have_birthday())
print(f"ë³€ê²½ëœ ì •ë³´: {person1.get_info()}")

print(f"\nìµœì¢… ì¸êµ¬: {Person.get_population()}ëª…")
````

````python
# ìƒì†ì—ì„œì˜ clsì™€ self ë™ì‘
class Animal:
    species_count = 0
    
    def __new__(cls, name):
        print(f"Animal.__new__ í˜¸ì¶œ, cls={cls.__name__}")
        cls.species_count += 1
        instance = super().__new__(cls)
        return instance
    
    def __init__(self, name):
        print(f"Animal.__init__ í˜¸ì¶œ, self íƒ€ì…={type(self).__name__}")
        self.name = name
    
    @classmethod
    def get_species_info(cls):
        return f"{cls.__name__} ì¢…ì˜ ê°œì²´ ìˆ˜: {cls.species_count}"
    
    def make_sound(self):
        return f"{self.name}ê°€ ì†Œë¦¬ë¥¼ ëƒ…ë‹ˆë‹¤."

class Dog(Animal):
    def __new__(cls, name, breed):
        print(f"Dog.__new__ í˜¸ì¶œ, cls={cls.__name__}")
        # ë¶€ëª¨ì˜ __new__ í˜¸ì¶œ ì‹œ clsê°€ Dogë¡œ ì „ë‹¬ë¨
        instance = super().__new__(cls, name)
        return instance
    
    def __init__(self, name, breed):
        print(f"Dog.__init__ í˜¸ì¶œ, self íƒ€ì…={type(self).__name__}")
        super().__init__(name)  # ë¶€ëª¨ì˜ __init__ í˜¸ì¶œ
        self.breed = breed
    
    def make_sound(self):
        return f"{self.name}({self.breed})ê°€ ë©ë© ì§–ìŠµë‹ˆë‹¤."

class Cat(Animal):
    def make_sound(self):
        return f"{self.name}ê°€ ì•¼ì˜¹ ì›ë‹ˆë‹¤."

print("=== ìƒì†ì—ì„œì˜ clsì™€ self ë™ì‘ ===")

# Dog ì¸ìŠ¤í„´ìŠ¤ ìƒì„±
dog = Dog("ë°”ë‘‘ì´", "ì§„ë—ê°œ")
print(f"Dog ìƒì„± ì™„ë£Œ: {dog.make_sound()}")

# Cat ì¸ìŠ¤í„´ìŠ¤ ìƒì„±  
cat = Cat("ë‚˜ë¹„")
print(f"Cat ìƒì„± ì™„ë£Œ: {cat.make_sound()}")

# í´ë˜ìŠ¤ ë©”ì„œë“œ í˜¸ì¶œ - clsê°€ ê°ê° ë‹¤ë¥´ê²Œ ì „ë‹¬ë¨
print(f"\n=== í´ë˜ìŠ¤ë³„ ì •ë³´ ===")
print(Animal.get_species_info())  # cls = Animal
print(Dog.get_species_info())     # cls = Dog (í•˜ì§€ë§Œ species_countëŠ” Animalì—ì„œ ê´€ë¦¬)
print(Cat.get_species_info())     # cls = Cat (í•˜ì§€ë§Œ species_countëŠ” Animalì—ì„œ ê´€ë¦¬)

# ì£¼ìš” í¬ì¸íŠ¸ í™•ì¸
print(f"\n=== í•µì‹¬ í¬ì¸íŠ¸ ===")
print(f"dogì˜ ì‹¤ì œ íƒ€ì…: {type(dog).__name__}")
print(f"dog.__class__: {dog.__class__.__name__}")
print(f"isinstance(dog, Animal): {isinstance(dog, Animal)}")
print(f"isinstance(dog, Dog): {isinstance(dog, Dog)}")
````

## 7. í•µì‹¬ ì •ë¦¬ ë° ì‹¤ì „ íŒ

### ğŸ¯ í•µì‹¬ ê°œë… ìš”ì•½

#### `__new__()` vs `__init__()`
- **`__new__()`**: ê°ì²´ë¥¼ **ìƒì„±**í•˜ëŠ” ì—­í•  (ë©”ëª¨ë¦¬ í• ë‹¹)
- **`__init__()`**: ìƒì„±ëœ ê°ì²´ë¥¼ **ì´ˆê¸°í™”**í•˜ëŠ” ì—­í• 

#### `cls` vs `self`
- **`cls`**: í´ë˜ìŠ¤ ìì²´ë¥¼ ì°¸ì¡° (í´ë˜ìŠ¤ ì°¨ì›ì˜ ì‘ì—…)
- **`self`**: ì¸ìŠ¤í„´ìŠ¤ë¥¼ ì°¸ì¡° (ì¸ìŠ¤í„´ìŠ¤ ì°¨ì›ì˜ ì‘ì—…)

### ğŸ”§ ì–¸ì œ ì‚¬ìš©í•˜ë‚˜?

| ìƒí™© | ì‚¬ìš©í•  ë©”ì„œë“œ | ë§¤ê°œë³€ìˆ˜ | ì˜ˆì‹œ |
|------|---------------|----------|------|
| ë¶ˆë³€ ê°ì²´ ìƒì† | `__new__()` | `cls` | `Float_Units(float)` |
| ì‹±ê¸€í†¤ íŒ¨í„´ | `__new__()` | `cls` | `Singleton` |
| ê°ì²´ ìƒì„± ì œì–´ | `__new__()` | `cls` | ì¸ìŠ¤í„´ìŠ¤ ê°œìˆ˜ ì œí•œ |
| ì¼ë°˜ì ì¸ ì´ˆê¸°í™” | `__init__()` | `self` | ëŒ€ë¶€ë¶„ì˜ í´ë˜ìŠ¤ |
| ëŒ€ì•ˆ ìƒì„±ì | `@classmethod` | `cls` | `Person.from_birth_year()` |

### âš ï¸ ì£¼ì˜ì‚¬í•­

1. **`__new__()`ì—ì„œ ë°˜ë“œì‹œ ì¸ìŠ¤í„´ìŠ¤ë¥¼ ë°˜í™˜í•´ì•¼ í•¨**
2. **`__new__()`ì—ì„œ `None`ì„ ë°˜í™˜í•˜ë©´ `__init__()`ì´ í˜¸ì¶œë˜ì§€ ì•ŠìŒ**
3. **ë¶ˆë³€ ê°ì²´ ìƒì† ì‹œ `__init__()`ë§Œìœ¼ë¡œëŠ” ë¶€ì¡±í•¨**
4. **ë©”íƒ€í´ë˜ìŠ¤ëŠ” ê³ ê¸‰ ì£¼ì œì´ë¯€ë¡œ í•„ìš”í•  ë•Œë§Œ ì‚¬ìš©**

### ğŸ’¡ ì‹¤ì „ íŒ

- 99%ì˜ ê²½ìš° `__init__()`ë§Œ ì‚¬ìš©í•˜ë©´ ì¶©ë¶„
- ë¶ˆë³€ ê°ì²´ë¥¼ ìƒì†í•  ë•Œë§Œ `__new__()` ê³ ë ¤
- í´ë˜ìŠ¤ ë³€ìˆ˜ë‚˜ í´ë˜ìŠ¤ ë©”ì„œë“œê°€ í•„ìš”í•˜ë©´ `@classmethod` ì‚¬ìš©
- ë³µì¡í•œ ê°ì²´ ìƒì„± ë¡œì§ì€ íŒ©í† ë¦¬ í•¨ìˆ˜ë¡œ ë¶„ë¦¬í•˜ëŠ” ê²ƒë„ ì¢‹ì€ ë°©ë²•

````python
# ğŸ¯ ì¢…í•© ì˜ˆì œ: ëª¨ë“  ê°œë…ì„ í™œìš©í•œ í´ë˜ìŠ¤
class SmartNumber(int):
    """ìˆ«ìë¥¼ í™•ì¥í•œ ìŠ¤ë§ˆíŠ¸ í´ë˜ìŠ¤ - ëª¨ë“  ê°œë… ì¢…í•©"""
    
    # í´ë˜ìŠ¤ ë³€ìˆ˜
    created_count = 0
    instances = []
    
    def __new__(cls, value, description=""):
        """clsë¥¼ ì‚¬ìš©í•œ ê°ì²´ ìƒì„± ì œì–´"""
        print(f"ğŸ”§ SmartNumber.__new__(cls={cls.__name__}, value={value})")
        
        # ë¶ˆë³€ ê°ì²´(int) ìƒì†ì´ë¯€ë¡œ __new__ì—ì„œ ê°’ ì„¤ì •
        if not isinstance(value, (int, float)):
            raise TypeError("ìˆ«ìë§Œ í—ˆìš©ë©ë‹ˆë‹¤")
        
        instance = super().__new__(cls, int(value))
        
        # í´ë˜ìŠ¤ ë³€ìˆ˜ ì—…ë°ì´íŠ¸ (cls ì‚¬ìš©)
        cls.created_count += 1
        cls.instances.append(instance)
        
        print(f"   ìƒì„±ëœ ì¸ìŠ¤í„´ìŠ¤: {instance} (ì´ {cls.created_count}ê°œ)")
        return instance
    
    def __init__(self, value, description=""):
        """selfë¥¼ ì‚¬ìš©í•œ ì¸ìŠ¤í„´ìŠ¤ ì´ˆê¸°í™”"""
        print(f"ğŸ¨ SmartNumber.__init__(self={self}, description='{description}')")
        
        # selfë¡œ ì¸ìŠ¤í„´ìŠ¤ ë³€ìˆ˜ ì„¤ì •
        self.description = description
        self.creation_order = self.__class__.created_count
        
    @classmethod
    def from_string(cls, text):
        """clsë¥¼ ì‚¬ìš©í•œ ëŒ€ì•ˆ ìƒì„±ì"""
        print(f"ğŸ“ from_string(cls={cls.__name__}, text='{text}')")
        
        # ë¬¸ìì—´ì—ì„œ ìˆ«ì ì¶”ì¶œ
        import re
        numbers = re.findall(r'\d+', text)
        if not numbers:
            raise ValueError("ë¬¸ìì—´ì—ì„œ ìˆ«ìë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤")
        
        value = int(numbers[0])
        return cls(value, f"'{text}'ì—ì„œ ì¶”ì¶œ")
    
    @classmethod
    def get_statistics(cls):
        """clsë¥¼ ì‚¬ìš©í•œ í´ë˜ìŠ¤ ì •ë³´ ì¡°íšŒ"""
        return {
            'total_count': cls.created_count,
            'instances': [int(inst) for inst in cls.instances],
            'sum': sum(cls.instances),
            'average': sum(cls.instances) / len(cls.instances) if cls.instances else 0
        }
    
    def get_info(self):
        """selfë¥¼ ì‚¬ìš©í•œ ì¸ìŠ¤í„´ìŠ¤ ì •ë³´"""
        return {
            'value': int(self),
            'description': self.description,
            'creation_order': self.creation_order,
            'type': type(self).__name__
        }
    
    def __str__(self):
        """selfë¥¼ ì‚¬ìš©í•œ ë¬¸ìì—´ í‘œí˜„"""
        if self.description:
            return f"{int(self)} ({self.description})"
        return str(int(self))

print("=" * 50)
print("ğŸš€ SmartNumber ì¢…í•© í…ŒìŠ¤íŠ¸")
print("=" * 50)

# ë‹¤ì–‘í•œ ìƒì„± ë°©ë²• í…ŒìŠ¤íŠ¸
print("\n1ï¸âƒ£ ì¼ë°˜ ìƒì„±ì ì‚¬ìš©")
num1 = SmartNumber(42, "ì •ë‹µ")
print(f"ê²°ê³¼: {num1}")

print("\n2ï¸âƒ£ ëŒ€ì•ˆ ìƒì„±ì ì‚¬ìš© (cls í™œìš©)")
num2 = SmartNumber.from_string("ì ìˆ˜ëŠ” 95ì ì…ë‹ˆë‹¤")
print(f"ê²°ê³¼: {num2}")

print("\n3ï¸âƒ£ ê°„ë‹¨í•œ ìƒì„±")
num3 = SmartNumber(100)
print(f"ê²°ê³¼: {num3}")

print("\n4ï¸âƒ£ ì¸ìŠ¤í„´ìŠ¤ ì •ë³´ í™•ì¸ (self í™œìš©)")
for i, num in enumerate([num1, num2, num3], 1):
    print(f"num{i} ì •ë³´: {num.get_info()}")

print("\n5ï¸âƒ£ í´ë˜ìŠ¤ í†µê³„ í™•ì¸ (cls í™œìš©)")
stats = SmartNumber.get_statistics()
print(f"í†µê³„: {stats}")

print("\n6ï¸âƒ£ intì²˜ëŸ¼ ì—°ì‚° ê°€ëŠ¥")
result = num1 + num2
print(f"{num1} + {num2} = {result} (íƒ€ì…: {type(result)})")

print("\nâœ¨ ëª¨ë“  ê°œë…ì´ ì„±ê³µì ìœ¼ë¡œ ì‘ë™í•©ë‹ˆë‹¤!")
````
